```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fromsklearn.datasets import samples_generator
from mpl_toolkits import mplot3d
```

```python
x, color = samples_generator.make_swiss_roll(n_samples=2000)
print('Shape of x', x.shape)
print('Shape of color', color.shape)
```

```python
x = pd.DataFrame(x)
ax= plt.subplots(figsize=(12,8))
ax =plt.axes(projection='3d')

ax.scatter3D(x[0], x[1], x[2], c=color, cmap=plt.cm.Spectral)
```

```python
def apply_manifold_learning(x, method):
    
    x = method.fit_transform(x)
    print('New Shape of X : ', x.shape)
    
    x = pd.DataFrame(x)
    
    plt.subplots(figsize=(12, 8))
    plt.axis('equal')
    
    plt.scatter(x[0], x[1], c= color, cmap =plt.cm.Spectral)
    plt.xlabel('x[0]')
    plt.ylabel('x[1]')
    
    return method
```

```python
# MDS  Multi Dimentional Scaling 

from sklearn.manifold import MDS

mds = apply_manifold_learning(x, MDS(n_components=2, metric=False))
```

```python
# MDS  Multi Dimentional Scaling 

from sklearn.manifold import MDS

mds = apply_manifold_learning(x, MDS(n_components=2, metric=True))
```

```python
# LocallyLinearEmbedding 
from sklearn.manifold import LocallyLinearEmbedding

lie = apply_manifold_learning(x, LocallyLinearEmbedding(
    n_neighbors = 15,
    n_components = 2,
    method = 'standard'
))
```

```python
# LocallyLinearEmbedding 
from sklearn.manifold import LocallyLinearEmbedding

lie = apply_manifold_learning(x, LocallyLinearEmbedding(
    n_neighbors = 15,
    n_components = 2,
    method = 'hessian'
))
```
